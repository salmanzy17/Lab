```python
for ojha, feature in enumerate(list(df_with_delimiter.columns)[:-1]):

    fg = sns.FacetGrid(df_with_delimiter, hue='Variety', height=5)

    fg.map(sns.displot, feature).add_legend()

    plt.show()
```

```python
sns.boxplot(x='Variety',y='petal length (cm)',data=df_with_delimiter)
plt.show()
```

```python
sns.violinplot(x='Variety',y='petal length (cm)',data=df_with_delimiter)
plt.show()
```

```python
df_with_delimiter.plot(kind='scatter',x='sepal length(cm)',y='sepal width (cm)', c= 'g')

plt.show()
```

```python
sns.set_style("whitegrid");

sns.FacetGrid(df_with_delimiter,hue="Variety", height=10).map(plt.scatter,"sepal length(cm)", "sepal width (cm)").add_legend();

plt.show();
```

```python
sns.set_style("whitegrid");

sns.pairplot(df_with_delimiter,vars =['sepal length(cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)','Variety'], hue = 'Variety', diag_kind = 'kde', plot_kws = {'alpha': 0.6, 's': 80, 'edgecolor': 'k'}, height = 2) 

plt.show()
```
